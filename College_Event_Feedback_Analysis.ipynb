```python
df.columns = [
    "Student_ID", "Subject_Knowledge", "Concept_Clarity", "Presentation_Use",
    "Assignment_Difficulty", "Doubt_Solving", "Course_Structure",
    "Student_Support", "Course_Recommendation"
]
```

```python
df["Satisfaction_Score"] = df[[
    "Subject_Knowledge", "Concept_Clarity", "Presentation_Use",
    "Assignment_Difficulty", "Doubt_Solving", "Course_Structure",
    "Student_Support", "Course_Recommendation"
]].mean(axis=1)
```

```python
top_3 = df.sort_values(by="Satisfaction_Score", ascending=False).head(3)
```

```python
top_3
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(df["Satisfaction_Score"], bins=10,kde=True,  color="skyblue")
plt.title("Distribution of Satisfaction Scores")
plt.xlabel("Satisfaction Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```

```python
average_ratings = df[[
    "Subject_Knowledge", "Concept_Clarity", "Presentation_Use",
    "Assignment_Difficulty", "Doubt_Solving", "Course_Structure",
    "Student_Support", "Course_Recommendation"
]].mean()
```

```python

plt.figure(figsize=(10, 8))
plt.pie(
    average_ratings,
    labels=average_ratings.index,
    autopct='%1.1f%%',
    startangle=40,
    colors=sns.color_palette("Set3")
)
plt.title("Average Ratings by Feedback Category")
plt.show()
```

```python
correlation_matrix = df.drop(columns=["Student_ID"]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Between Feedback Categories")
plt.show()
```

```python
sns.scatterplot(x="Assignment_Difficulty", y="Satisfaction_Score", data=df)
plt.title("Assignment Difficulty vs Satisfaction Score")
plt.xlabel("Assignment Difficulty")
plt.ylabel("Satisfaction Score")
plt.show()
```

```python
correlations = df.corr()["Satisfaction_Score"].drop(["Student_ID", "Satisfaction_Score"]).sort_values(ascending=False)
print(correlations)
```

```python
rating_columns = [
    "Subject_Knowledge", "Concept_Clarity", "Presentation_Use",
    "Assignment_Difficulty", "Doubt_Solving", "Course_Structure",
    "Student_Support", "Course_Recommendation"
]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[rating_columns], palette="Set2")
plt.xticks(rotation=45)
plt.title("Box Plot of Feedback Ratings")
plt.ylabel("Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(11,6))
rating_columns_with_satisfaction = rating_columns + ["Satisfaction_Score"]
sns.pairplot(df[rating_columns_with_satisfaction])
plt.title("Pair Plot of Feedback Ratings and Satisfaction Score")
plt.show()
```
